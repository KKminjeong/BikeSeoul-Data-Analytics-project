```python
df_rent = pd.read_csv('공공자전거 대여이력 정보_2020.10_UTF8.csv', encoding='utf-8')

df_rent
```

```python
df_rent = df_rent.drop((['자전거번호', '대여거치대', '반납일시', '반납대여소번호', '반납대여소명', '반납거치대']), axis = 1)

# drop을 이용해 불필요한 컬럼삭제

df_rent.head()
```

```python
df_rent = df_rent[df_rent['이용거리'] != 0] #이용거리가 0인 값 제거

df_rent.sort_values(by = ['대여일시']).head() #merge() 확인을 위한 정렬
```

```python
df_center = pd.read_csv('공공자전거 대여소 정보(21.01.31 기준)_UTF8.csv', encoding='utf-8')

df_center
```

```python
df_center = df_center.drop((['설치시기', '설치형태', '공란', '운영방식']), axis = 1)

df_center.head()
```

```python
# 대여기록 data와 대여소 정보 merge 진행

df_final = pd.merge(df_rent, df_center, 
                    left_on = '대여 대여소번호', right_on = '대여소번호', how ='inner')

df_final.sort_values(by=['대여일시']) #대여일시 열로 정렬 -> 전처리 결과 원본 데이터와 비교대조
```

```python
#자치구 별 대여 횟수 및 점포수량 집계
group_base = df_final.groupby('자치구')  #자치구별로 데이터 값을 묶어 집계
group_gu = group_base.size() # 배열의 모양과 크기 확인

df_gu = pd.DataFrame(group_gu, columns=['대여횟수'])


group_center = df_center.groupby('자치구') 
group_center_gu = group_center.size() 

df_group_center = pd.DataFrame(group_center_gu, columns=['점포수량'])


df_group_final = pd.merge(df_gu, df_group_center,                            #자치구 열을 기준으로 데이터프레임 병합
                    left_on = '자치구', right_on = '자치구', how ='inner')

df_group_final
```

### 2. 구별 대여정보 시각화

#### 2-1 구별 대여정보 지도 시각화

```python
# 자치구 별 대여횟수에 따른 시각화 자료 출력

import folium

import json

geo_path = 'skorea_municipalities_geo_simple.json'     
geo_str = json.load(open(geo_path, encoding='utf-8'))

# geo_path라는 변수에 json 파일 경로를 담는다


map1 = folium.Map(location=[37.5502, 126.982], zoom_start=11, tiles='Stamen Toner') 

# 타일 옵션을 사용하여 지도 생성
# zoom_start를 이용하여 확대 비율 조정


map1.choropleth(geo_data = geo_str,             
               data = df_gu['대여횟수'], 
               columns = [df_gu.index, df_gu['대여횟수']],
               fill_color = 'YlGn',
               key_on = 'feature.id')


# folium에서 choropleth 명령으로 json 파일과 지도에 표현하고 싶은 데이터를 입력
# key_on 옵션으로 지도의 id를 알려줌


map1
```

#### 2-2 구별 대여소 위치정보 시각화

```python
#상위 50개 대여소 위도 경도 data 수집

df_mark = df_final.groupby(['대여소번호','대여 대여소명', '자치구', '위도', '경도'])
group_sort = df_mark.size().sort_values(ascending = False)
df_info = pd.DataFrame(group_sort, columns=['대여횟수']).reset_index()
df_mark_top50 = df_info[:50]
```

```python
#자치구 시각화 자료 + 상위 50개 대여소 마킹

map2 = folium.Map(location=[37.5502, 126.982], zoom_start=11, tiles='Stamen Toner')
for i in range(len(df_mark_top50)):
    marker = folium.Marker(
    location=[df_mark_top50['위도'][i], df_mark_top50['경도'][i]],
    popup=folium.Popup(df_mark_top50['대여 대여소명'][i], max_width='100'))
    map2.add_child(marker)                # 지도를 그리고, 그 상태에서 원하는 좌표에 Marker 명령으로 마크를 찍음 
    i += 1

  
        
map2.choropleth(geo_data = geo_str, 
               data = df_gu['대여횟수'], 
               columns = [df_gu.index, df_gu['대여횟수']], 
               fill_color = 'YlGn',
               key_on = 'feature.id') 

map2
```
